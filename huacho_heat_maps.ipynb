```python
%pylab inline
import pandas as pnd
```

```python
huacho = pnd.read_csv('huacho.txt', skiprows=2, header='infer', index_col=0, sep='\t', parse_dates=True)
```

```python
print 'end date   =', huacho.index.max()
print 'start date =', huacho.index.min()

ds = '2012-11-01'
de = '2013-01-08'
```

```python
for i, col in enumerate(huacho.columns):
    print i, col
```

```python
for i in range(len(huacho.columns)):
    values = huacho[huacho.columns[i]][ds:de].values
    mapped = values.reshape((48, len(values)/48), order='F')
    imshow(mapped)
    title(huacho.columns[i])
    xlabel('date')
    ylabel('time of day')
    plt.show()
```
